```python
# совмещаем столбцы
df=df[df.id!='id']
```

```python
df["опыт'"].value_counts()
```

```python
df[10:20]
```

```python
# 2. Построение ящичных диаграмм для опыта и зарплаты
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x="опыт'", data=df)
plt.title('Ящик опыта')

plt.subplot(1, 2, 2)
sns.boxplot(x='зп', data=df)
plt.title('Ящик зарплаты')

plt.tight_layout()
plt.show()
```

```python
# 3. Проверка зависимости зарплаты от опыта
sns.scatterplot(x="опыт'", y='зп', data=df)
plt.title('Зависимость зарплаты от опыта')
plt.xlabel('Опыт')
plt.ylabel('Зарплата')
plt.show()
```

```python
# 5. Разделение вакансий на seniors и others
df['категория'] = df['описание'].apply(lambda x: 'seniors' if 'senior' in x.lower() else 'others')
```

```python
# 6. Проверка зарплат для обеих групп
plt.figure(figsize=(8, 6))
sns.boxplot(x='категория', y='зп', data=df)
plt.title('Сравнение зарплат для seniors и others')
plt.xlabel('Категория')
plt.ylabel('Зарплата')
plt.show()
```
